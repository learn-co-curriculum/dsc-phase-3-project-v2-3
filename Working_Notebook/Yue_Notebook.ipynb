```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier, DummyRegressor
import numpy as np
```

```python
df = pd.read_csv('../data/bigml_59c28831336c6604c800002a.csv')
df
```

```python
df = df.drop(columns = ['phone number', 'area code'])
```

```python
df['churn'].value_counts()
```

```python
df.info()
```

```python
df.head()
```

```python
df['average_hourly_night_rate'] = df['total night minutes'] / df['total night charge']
```

```python
df['average_hourly_day_rate'] = df['total day minutes'] / df['total day charge']
```

```python
df['average_hourly_intl_rate'] = df['total intl minutes'] / df['total intl charge']
```

```python
df['average_hourly_eve_rate'] = df['total eve minutes'] / df['total eve charge']
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.info()
```

```python
X = df.drop('churn', axis=1)
y = df['churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
train_state_cat = X_train['state']
```

```python
train_plan_cat = X_train[['international plan', 'voice mail plan']]
```

```python
train_plan_cat
```

## One Hot Encoding State Column

```python
ohe = OneHotEncoder(categories = 'auto', sparse=False, handle_unknown="ignore")
```

```python
train_state_encoded = ohe.fit_transform(np.array(train_state_cat).reshape(-1,1))
```

```python
ohe.categories_
```

```python
ohe.get_feature_names()
```

```python
train_state_encoded
```

```python
train_state_encoded = pd.DataFrame(
                        train_state_encoded,
                        columns=ohe.get_feature_names(), 
                        index=X_train.index,
)

train_state_encoded
```

```python
train_state_encoded.columns = train_state_encoded.columns.str.replace('x0_', '')
train_state_encoded
```

```python
test_state_cat =  X_test['state']
```

```python
test_plan_cat = X_test[['international plan', 'voice mail plan']]
```

```python
test_state_encoded = ohe.transform(np.array(test_state_cat).reshape(-1,1))
test_state_encoded
```

```python
test_state_encoded = pd.DataFrame(
                        test_state_encoded,
                        columns=ohe.get_feature_names(), 
                        index=X_test.index,
)

test_state_encoded
```

```python
test_state_encoded.columns = test_state_encoded.columns.str.replace('x0_', '')
test_state_encoded
```

## Ordinal Encoding Plan Columns

```python
ordi_encoder = OrdinalEncoder()
train_plan_encoded = ordi_encoder.fit_transform(train_plan_cat)
```

```python
train_plan_encoded.shape
```

```python
train_plan_encoded = pd.DataFrame(
                        train_plan_encoded,
                        columns=[('international plan'), ('voice mail plan')], 
                        index=X_train.index,
)

train_plan_encoded
```

```python
test_plan_encoded = ordi_encoder.transform(test_plan_cat)
test_plan_encoded.shape
```

```python
test_plan_encoded = pd.DataFrame(
                        test_plan_encoded,
                        columns=[('international plan'), ('voice mail plan')], 
                        index=X_test.index,
)

test_plan_encoded
```

## Training Data

```python
X_train.head()
```

```python
X_train.drop(columns=['state','international plan', 'voice mail plan'], inplace=True)
```

```python
training_data = pd.concat([X_train, train_state_encoded, train_plan_encoded, y_train], axis=1)
#training_data
training_data.to_csv('training_data.csv')
```

## Testing Data

```python
X_test.drop(columns=['state', 'international plan', 'voice mail plan'], inplace=True)
```

```python
testing_data = pd.concat([X_test, test_state_encoded, test_plan_encoded, y_test], axis=1)
# testing_data
testing_data.to_csv('testing_data.csv')
```
